```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
import pandas as pd
import time
```

[Dokumentation](http://selenium-python.readthedocs.io/)

```python
driver = webdriver.Firefox()
search_url = "https://dnszensur.ch/censorship_phishing.php"
driver.get(search_url)
time.sleep(2)
text = (driver.page_source)
soup = BeautifulSoup(text, 'html.parser')
```

```python
soup
```

```python
soup.find_all('td', {'style': ''})
```

```python
lst = []
for elem in soup.find_all('td', {'style': ''}):
    lst.append(elem.text.strip())
```

```python
lst
```

```python
lstnew = []
for x in lst:
    if x != '':
        lstnew.append(x)
```

```python
lst_dropped = [x for x in lst if x != '']
```
