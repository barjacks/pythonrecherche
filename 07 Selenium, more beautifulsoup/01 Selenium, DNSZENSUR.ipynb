```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
import pandas as pd
```

[Dokumentation](http://selenium-python.readthedocs.io/)

```python
driver = webdriver.Firefox()
search_url = "https://dnszensur.ch/censorship_phishing.php"
driver.get(search_url)
```

```python
text = (driver.page_source)
```

```python
soup = BeautifulSoup(text, 'html.parser')
```

```python
soup
```

```python
lst = []
for elem in soup.find_all('td', {'style': ''}):
    lst.append(elem.text.strip())
```

```python
lst_dropped = [x for x in lst if x != '']
```
