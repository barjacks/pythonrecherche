# Imports

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# URL Lesen

```python
url = "https://www.iaaf.org/records/toplists/sprints/100-metres/outdoor/men/senior/2017"
response = requests.get(url)
sprinters_soup = BeautifulSoup(response.text, 'html.parser')
```

# URL Code der gesamten Site

```python
#sprinters_soup um den Code anzuzeigen den Hashtag entfernen
```

# URL Code eingrenzen 

```python
lst = sprinters_soup.find_all('table', {'class':'records-table'})
```

# Listenlänge prüfen

```python
len(lst)
```

# Relevanten URL Code finden

```python
sprinterliste = lst[2]
```

# Relevanten URL Code anzeigen

```python
# sprinterliste um den Code anzuzeigen, den Hashtag entfernen.
```

# Headers auslesen

```python
headers = []
for elem in sprinterliste.find_all('th'):
    headers.append(elem.text.strip())
```

```python
headers
```

# List Comprehension (For loop abgekürzt)

```python
[x for x in headers if x != '']
```

# Namen heraus filtern

```python
namens_list = sprinterliste.find_all('td', {'data-th':'Competitor'})
```

```python
type(namens_list)
```

```python
len(namens_list)
```

```python
namen = []
for elem in namens_list:
    namen.append(elem.text.strip())
```

```python
namen
```

# IDs aus HREF links extrahieren

```python
id_list = sprinterliste.find_all('td', {'data-th':'Competitor'})
```

```python
ids = []
for elem in id_list:
    ids.append(elem.find('a')['href'].replace('/athletes/athlete=', ''))
```

```python
ids
```

# Dataframe bauen mit ID und Namen

```python
df = pd.DataFrame(namen)
df['ID'] = ids
df.columns = [['Name', 'ID']]
```
