```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.iaaf.org/records/toplists/sprints/100-metres/outdoor/men/senior/2017"
response = requests.get(url)
sprinters_soup = BeautifulSoup(response.text, 'html.parser')
```

```python
lst = sprinters_soup.find_all('table', {'class':'records-table'})
```
